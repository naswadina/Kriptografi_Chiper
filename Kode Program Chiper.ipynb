```python
import tkinter as tk  # Mengimpor tkinter untuk GUI
from tkinter import filedialog, messagebox  # Mengimpor dialog file dan kotak pesan
import string  # Mengimpor modul string untuk penanganan karakter
import numpy as np  # Mengimpor NumPy untuk operasi matriks
from math import gcd  # Mengimpor fungsi gcd untuk perhitungan matematis

# Implementasi Vigenere Cipher
def vigenere_enskripsi(teks_biasa, kunci):
    teks_biasa = ''.join(filter(str.isalpha, teks_biasa)).upper()  # Menyaring dan mengubah teks biasa menjadi huruf besar
    kunci = kunci.upper()  # Mengubah kunci menjadi huruf besar
    sandi = ''.join(  # Enkripsi teks
        chr(((ord(char) - ord('A') + (ord(kunci[i % len(kunci)]) - ord('A'))) % 26) + ord('A'))
        for i, char in enumerate(teks_biasa)
    )
    return sandi  # Mengembalikan teks terenkripsi

def vigenere_dekripsi(teks_sandi, kunci):
    teks_sandi = ''.join(filter(str.isalpha, teks_sandi)).upper()  # Menyaring dan mengubah teks sandi menjadi huruf besar
    kunci = kunci.upper()  # Mengubah kunci menjadi huruf besar
    teks_biasa = ''.join(  # Dekripsi teks
        chr(((ord(char) - ord('A') - (ord(kunci[i % len(kunci)]) - ord('A'))) % 26) + ord('A'))
        for i, char in enumerate(teks_sandi)
    )
    return teks_biasa  # Mengembalikan teks asli

# Implementasi Playfair Cipher
def generate_playfair_matrix(kunci):
    kunci = ''.join(dict.fromkeys(kunci.upper().replace('J', 'I')))  # Menghasilkan matriks kunci
    matriks = []
    for char in kunci:
        if char not in matriks and char in string.ascii_uppercase:
            matriks.append(char)  # Menambahkan karakter unik dari kunci
    for char in string.ascii_uppercase:
        if char == 'J':
            continue  # Mengabaikan 'J'
        if char not in matriks:
            matriks.append(char)  # Mengisi sisa alfabet
    return [matriks[i:i+5] for i in range(0, 25, 5)]  # Mengembalikan matriks 5x5

def cari_posisi(matriks, char):
    for baris in range(5):
        for kolom in range(5):
            if matriks[baris][kolom] == char:
                return baris, kolom  # Mengembalikan posisi karakter dalam matriks
    return None

def playfair_enskripsi(teks_biasa, kunci):
    matriks = generate_playfair_matrix(kunci)  # Membuat matriks Playfair
    teks_biasa = ''.join(filter(str.isalpha, teks_biasa.upper())).replace('J', 'I')  # Menyaring dan mengganti 'J' dengan 'I'

    pasangan = []  # Untuk menyimpan pasangan huruf
    i = 0
    while i < len(teks_biasa):
        a = teks_biasa[i]  # Mengambil huruf pertama
        b = ''
        if i + 1 < len(teks_biasa):
            b = teks_biasa[i + 1]  # Mengambil huruf kedua
            if a == b:  # Jika pasangan sama
                b = 'X'  # Ganti dengan 'X'
                i += 1
            else:
                i += 2  # Melanjutkan ke dua huruf berikutnya
        else:
            b = 'X'  # Jika hanya ada satu huruf, ganti dengan 'X'
            i += 1
        pasangan.append(a + b)  # Menyimpan pasangan huruf

    sandi = ''
    for pair in pasangan:
        baris1, kolom1 = cari_posisi(matriks, pair[0])  # Mencari posisi huruf pertama
        baris2, kolom2 = cari_posisi(matriks, pair[1])  # Mencari posisi huruf kedua
        if baris1 == baris2:
            sandi += matriks[baris1][(kolom1 + 1) % 5]  # Sama baris, geser ke kanan
            sandi += matriks[baris2][(kolom2 + 1) % 5]
        elif kolom1 == kolom2:
            sandi += matriks[(baris1 + 1) % 5][kolom1]  # Sama kolom, geser ke bawah
            sandi += matriks[(baris2 + 1) % 5][kolom2]
        else:
            sandi += matriks[baris1][kolom2]  # Berbeda baris dan kolom, tukar
            sandi += matriks[baris2][kolom1]
    return sandi  # Mengembalikan teks terenkripsi

def playfair_dekripsi(teks_sandi, kunci):
    matriks = generate_playfair_matrix(kunci)  # Membuat matriks Playfair
    teks_sandi = ''.join(filter(str.isalpha, teks_sandi.upper())).replace('J', 'I')  # Menyaring dan mengganti 'J' dengan 'I'

    pasangan = [teks_sandi[i:i + 2] for i in range(0, len(teks_sandi), 2)]  # Membuat pasangan dari teks sandi
    teks_biasa = ''
    for pair in pasangan:
        baris1, kolom1 = cari_posisi(matriks, pair[0])  # Mencari posisi huruf pertama
        baris2, kolom2 = cari_posisi(matriks, pair[1])  # Mencari posisi huruf kedua
        if baris1 == baris2:
            teks_biasa += matriks[baris1][(kolom1 - 1) % 5]  # Sama baris, geser ke kiri
            teks_biasa += matriks[baris2][(kolom2 - 1) % 5]
        elif kolom1 == kolom2:
            teks_biasa += matriks[(baris1 - 1) % 5][kolom1]  # Sama kolom, geser ke atas
            teks_biasa += matriks[(baris2 - 1) % 5][kolom2]
        else:
            teks_biasa += matriks[baris1][kolom2]  # Berbeda baris dan kolom, tukar
            teks_biasa += matriks[baris2][kolom1]
    return teks_biasa  # Mengembalikan teks asli

# Implementasi Hill Cipher
def invers_matriks_mod(matriks, mod):
    det = int(round(np.linalg.det(matriks)))  # Menghitung determinan matriks
    det = det % mod  # Mengambil determinan modulo
    inv_det = pow(det, -1, mod)  # Mencari invers determinan
    if inv_det is None:
        return None  # Jika tidak ada invers, kembalikan None

    cofactors = np.linalg.inv(matriks) * det  # Menghitung kofaktor
    inv_matriks = (inv_det * cofactors) % mod  # Menghitung invers matriks
    return np.round(inv_matriks).astype(int)  # Mengembalikan matriks invers

def hill_cipher(text, key, mode):
    key_matrix = np.array([ord(c) - ord('A') for c in key.upper()[:9]]).reshape(3, 3)  # Membentuk matriks kunci 3x3
    text = text.upper()  # Mengubah teks menjadi huruf besar
    result = ""

    # Memastikan panjang teks adalah kelipatan dari 3
    text = text.ljust((len(text) + 2) // 3 * 3, 'X')  # Menambahkan 'X' jika perlu

    for i in range(0, len(text), 3):
        chunk = text[i:i+3]  # Mengambil 3 huruf
        chunk_vector = np.array([ord(c) - ord('A') for c in chunk])  # Mengubah huruf menjadi vektor

        if mode == "Enkripsi":
            encrypted_vector = np.dot(key_matrix, chunk_vector) % 26  # Mengalikan matriks kunci untuk enkripsi
        else:  # mode == "Dekripsi"
            det = int(round(np.linalg.det(key_matrix))) % 26  # Menghitung determinan
            inv_det = pow(det, -1, 26)  # Mencari invers determinan
            key_matrix_inv = (inv_det * np.round(np.linalg.inv(key_matrix) * det) % 26).astype(int)  # Menghitung matriks kunci invers
            encrypted_vector = np.dot(key_matrix_inv, chunk_vector) % 26  # Mengalikan matriks kunci invers untuk dekripsi

        result += ''.join([chr(int(v) + ord('A')) for v in encrypted_vector])  # Mengubah hasil kembali menjadi huruf

    return result  # Mengembalikan hasil

# GUI dengan tkinter
class AplikasiKripto:
    def __init__(self, master):
        self.master = master
        master.title("Kriptografi")  # Judul aplikasi

        # Label dan Input Teks
        self.label_input_teks = tk.Label(master, text="Input Teks:")
        self.label_input_teks.grid(row=0, column=0, padx=10, pady=10, sticky='w')

        self.input_teks = tk.Text(master, height=5, width=40)  # Area input untuk teks
        self.input_teks.grid(row=1, column=0, columnspan=3, padx=10, pady=5)

        # Tombol Unggah File
        self.tombol_unggah = tk.Button(master, text="Unggah File", command=self.unggah_file, width=12)  # Tombol untuk unggah file
        self.tombol_unggah.grid(row=1, column=3, padx=10, pady=5)

        # Label dan Entry Kunci
        self.label_kunci = tk.Label(master, text="Kunci (min 12 karakter):")
        self.label_kunci.grid(row=2, column=0, padx=10, pady=10, sticky='w')

        self.entry_kunci = tk.Entry(master, width=30)  # Entry untuk kunci
        self.entry_kunci.grid(row=2, column=1, padx=10, pady=10, sticky='w')

        # Dropdown untuk memilih Cipher
        self.label_cipher = tk.Label(master, text="Pilih Cipher:")
        self.label_cipher.grid(row=3, column=0, padx=10, pady=10, sticky='w')

        self.var_cipher = tk.StringVar(master)
        self.var_cipher.set("Vigenere")  # Menetapkan cipher default
        self.menu_cipher = tk.OptionMenu(master, self.var_cipher, "Vigenere", "Playfair", "Hill")  # Menu pilihan cipher
        self.menu_cipher.grid(row=3, column=1, padx=10, pady=10, sticky='w')

        # Tombol Enkripsi dan Dekripsi
        self.tombol_enskripsi = tk.Button(master, text="Enkripsi", command=self.enkripsi, width=10)  # Tombol untuk enkripsi
        self.tombol_enskripsi.grid(row=4, column=0, padx=5, pady=5)

        self.tombol_dekripsi = tk.Button(master, text="Dekripsi", command=self.dekripsi, width=10)  # Tombol untuk dekripsi
        self.tombol_dekripsi.grid(row=4, column=1, padx=5, pady=5)

        # Label dan Output Hasil
        self.label_hasil = tk.Label(master, text="Hasil:")
        self.label_hasil.grid(row=5, column=0, padx=10, pady=10, sticky='w')

        self.output_hasil = tk.Text(master, height=5, width=40)  # Area output untuk hasil
        self.output_hasil.grid(row=6, column=0, columnspan=3, padx=10, pady=5)

    def unggah_file(self):
        file_path = filedialog.askopenfilename()  # Dialog untuk memilih file
        with open(file_path, 'r') as file:
            teks = file.read()  # Membaca isi file
            self.input_teks.delete(1.0, tk.END)  # Menghapus teks yang ada
            self.input_teks.insert(tk.END, teks)  # Memasukkan teks dari file

    def enkripsi(self):
        teks_biasa = self.input_teks.get(1.0, tk.END).strip()  # Mengambil teks untuk dienkripsi
        kunci = self.entry_kunci.get().strip()  # Mengambil kunci dari input
        cipher = self.var_cipher.get()  # Mendapatkan jenis cipher yang dipilih

        if len(kunci) < 12:  # Validasi panjang kunci
            messagebox.showerror("Error", "Kunci harus minimal 12 karakter.")
            return

        if not kunci.isalpha():  # Validasi kunci hanya huruf
            messagebox.showerror("Error", "Kunci harus hanya terdiri dari huruf alfabet.")
            return

        try:
            if cipher == "Vigenere":
                hasil = vigenere_enskripsi(teks_biasa, kunci)  # Enkripsi dengan Vigenere
            elif cipher == "Playfair":
                hasil = playfair_enskripsi(teks_biasa, kunci)  # Enkripsi dengan Playfair
            elif cipher == "Hill":
                hasil = hill_cipher(teks_biasa, kunci, "Enkripsi")  # Enkripsi dengan Hill
            else:
                hasil = "Cipher tidak dikenali."
            self.output_hasil.delete(1.0, tk.END)  # Menghapus hasil yang ada
            self.output_hasil.insert(tk.END, hasil)  # Menampilkan hasil enkripsi
        except Exception as e:
            messagebox.showerror("Error", str(e))  # Menampilkan pesan error jika terjadi

    def dekripsi(self):
        teks_sandi = self.input_teks.get(1.0, tk.END).strip()  # Mengambil teks untuk didekripsi
        kunci = self.entry_kunci.get().strip()  # Mengambil kunci dari input
        cipher = self.var_cipher.get()  # Mendapatkan jenis cipher yang dipilih

        if len(kunci) < 12:  # Validasi panjang kunci
            messagebox.showerror("Error", "Kunci harus minimal 12 karakter.")
            return

        if not kunci.isalpha():  # Validasi kunci hanya huruf
            messagebox.showerror("Error", "Kunci harus hanya terdiri dari huruf alfabet.")
            return

        try:
            if cipher == "Vigenere":
                hasil = vigenere_dekripsi(teks_sandi, kunci)  # Dekripsi dengan Vigenere
            elif cipher == "Playfair":
                hasil = playfair_dekripsi(teks_sandi, kunci)  # Dekripsi dengan Playfair
            elif cipher == "Hill":
                hasil = hill_cipher(teks_sandi, kunci, "Dekripsi")  # Dekripsi dengan Hill
            else:
                hasil = "Cipher tidak dikenali."
            self.output_hasil.delete(1.0, tk.END)  # Menghapus hasil yang ada
            self.output_hasil.insert(tk.END, hasil)  # Menampilkan hasil dekripsi
        except Exception as e:
            messagebox.showerror("Error", str(e))  # Menampilkan pesan error jika terjadi

if __name__ == "__main__":
    root = tk.Tk()  # Membuat jendela utama
    app = AplikasiKripto(root)  # Inisialisasi aplikasi
    root.mainloop()  # Menjalankan loop utama GUI
```
